```python
from flask import Flask, request, jsonify
# from flask_ngrok import run_with_ngrok
from transformers import AutoTokenizer, pipeline
import torch
import transformers
```

```python
model = "codellama/CodeLlama-7b-Instruct-hf"
```

```python

tokenizer = AutoTokenizer.from_pretrained(model)
text_generator = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
class ModelInput(BaseModel):
    Query: str
```

```python
from fastapi.responses import JSONResponse
# app = Flask(__name__)
# run_with_ngrok(app)

@app.post("/generate")
async def generate(input_parameters: ModelInput):
    user_input = input_parameters.Query
    prompt = f"<s><<SYS>>\nProvide answers as the user asks\n<</SYS>>\n\n{user_input}"
    sequences = text_generator(
        prompt,
        do_sample=True,
        top_k=10,
        temperature=0.1,
        top_p=0.95,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=250,
        add_special_tokens=False
    )
    return JSONResponse(content={'response': sequences[0]['generated_text']})

```
